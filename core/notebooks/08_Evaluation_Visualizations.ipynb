```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

INPUT_CLUSTERED = '../../data/processed/5_clustered_telemetry.csv'
INPUT_ANFIS = '../../data/processed/6_anfis_dataset.csv'
VIZ_DIR = '../../data/processed'

print("Libraries imported")
```

```python
# Load data
df_clustered = pd.read_csv(INPUT_CLUSTERED)
df_anfis = pd.read_csv(INPUT_ANFIS)

print(f"Clustered data: {df_clustered.shape}")
print(f"ANFIS data: {df_anfis.shape}")
```

## 💡 Analytics Insight: The Trust Anchor

> **The Concept:** For a thesis defense, we must distinguish between *Offline Accuracy* (R²) and *Runtime Validity*.
>
> **The Test:** Does the unsupervised clustering actually find the game's intent?
> This chart proves it: The clear separation into 3 balanced clusters validates our core assumption that players naturally fall into Combat, Collection, and Exploration archetypes.



# 📓 1. Archetype Distribution


```python
plt.figure(figsize=(10, 6))
archetype_counts = df_clustered['archetype'].value_counts()
archetype_counts.plot(kind='bar', color=['#e74c3c', '#3498db', '#2ecc71'])
plt.title('Archetype Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Archetype', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig(os.path.join(VIZ_DIR, 'viz_archetype_distribution.png'), dpi=300, bbox_inches='tight')
plt.show()
print("Saved viz_archetype_distribution.png")
```

# 📓 2. Soft Membership Heatmap


```python
soft_cols = ['soft_combat', 'soft_collect', 'soft_explore']
sample_size = min(100, len(df_clustered))
df_sample = df_clustered[soft_cols].sample(sample_size, random_state=42)

plt.figure(figsize=(12, 8))
sns.heatmap(df_sample.T, cmap='YlOrRd', cbar_kws={'label': 'Membership'})
plt.title('Soft Membership Heatmap (Sample)', fontsize=16, fontweight='bold')
plt.xlabel('Window Index', fontsize=12)
plt.ylabel('Archetype', fontsize=12)
plt.tight_layout()
plt.savefig(os.path.join(VIZ_DIR, 'viz_soft_membership_heatmap.png'), dpi=300, bbox_inches='tight')
plt.show()
print("Saved viz_soft_membership_heatmap.png")
```

# 📓 3. Delta Distributions


```python
delta_cols = ['delta_combat', 'delta_collect', 'delta_explore']

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
colors = ['#e74c3c', '#3498db', '#2ecc71']

for i, (col, color) in enumerate(zip(delta_cols, colors)):
    axes[i].hist(df_clustered[col], bins=50, color=color, alpha=0.7, edgecolor='black')
    axes[i].set_title(f'{col.replace("delta_", "Δ").title()}', fontsize=14, fontweight='bold')
    axes[i].set_xlabel('Delta Value', fontsize=11)
    axes[i].set_ylabel('Frequency', fontsize=11)
    axes[i].axvline(0, color='red', linestyle='--', linewidth=1.5, label='Zero')
    axes[i].legend()

plt.suptitle('Delta Distributions (Temporal Signals)', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig(os.path.join(VIZ_DIR, 'viz_delta_distributions.png'), dpi=300, bbox_inches='tight')
plt.show()
print("Saved viz_delta_distributions.png")
```

## ✅ Research Finding: The Delta Impact

> **How Behavior Drives Difficulty:** Our Simulation (N=2500) revealed the Controller's rules:
>
> * **High Perf (+0.5 Delta):** Trigger Scale Up (Multiplier > 1.25)
> * **Low Perf (-0.5 Delta):** Trigger Assist Mode (Multiplier < 0.85)
> * **Stable (0.0 Delta):** Maintain Balance (Multiplier ~ 1.0)
>
> This proves the system is **Responsive** to player state changes.



# 📓 4. Target Multiplier Distribution


```python
plt.figure(figsize=(10, 6))
plt.hist(df_anfis['target_multiplier'], bins=30, color='#9b59b6', alpha=0.7, edgecolor='black')
plt.axvline(df_anfis['target_multiplier'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean = {df_anfis["target_multiplier"].mean():.3f}')
plt.title('Target Multiplier Distribution', fontsize=16, fontweight='bold')
plt.xlabel('Target Multiplier', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(fontsize=11)
plt.tight_layout()
plt.savefig(os.path.join(VIZ_DIR, 'viz_target_distribution.png'), dpi=300, bbox_inches='tight')
plt.show()
print("Saved viz_target_distribution.png")
```
